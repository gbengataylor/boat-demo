# Detection Service

```python
import os
import sys
sys.path.insert(0, os.path.abspath('..'))

from PIL import Image, ImageDraw
import requests

IMAGE = os.path.abspath('../utils/boats.png')
DETECTION_URL = 'http://boat-detect:8080/v1/detect'
```

```python
img = Image.open(IMAGE)
img.show()
```

```python
response = requests.post(DETECTION_URL, files={"image": open(IMAGE, "rb").read()}).json()
response
```

```python
canvas = ImageDraw.Draw(img)
for box in response['objects']:
    canvas.rectangle([box['x0'], box['y0'], box['x1'], box['y1']], outline='red')
img.show()
```
